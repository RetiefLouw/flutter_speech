```python
import numpy as np
import torch
from pathlib import Path
```

```python
# features_dir = "out/wav2vec2-large-xlsr-53-layer12-mvn/npy"
features_dir = "temp_adult-mvn/npy/"
out_fn = f"{Path(features_dir).parent}/embedings.npz"
print(out_fn)
```

```python
z_dict = {}

for npy_fn in Path(features_dir).rglob("*.npy"):

    npy = np.load(npy_fn)
    z = np.mean(npy, axis=0)
    z_dict[npy_fn.stem] = z

np.savez(out_fn, **z_dict)
```
